##Лінійні графіки

Дані кліматичних спостережень https://www.kaggle.com/datasets/sumanthvrao/daily-climate-time-series-data

```python
climate_df = pd.read_csv('/drive/My Drive/!Kafedra/GOIT/mod2/DailyDelhiClimateTrain.csv', parse_dates=['date'], index_col=['date'])
print(climate_df.head())
month_climate = climate_df.resample('M').mean()
print(month_climate.head())

print(climate_df.info())
print(climate_df.describe())


climate_df_2015 = climate_df['2015-01-01': '2016-01-01']
climate_df_2016 = climate_df['2016-01-01': '2017-01-01']
```

Простий лінійний графік

```python
plt.plot(month_climate.index, month_climate.meantemp, label='day temperature')
plt.show()
```

Додавання підписів вісям та назви графіку

```python
plt.plot(month_climate.index, month_climate.meantemp, label='day temperature')
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Температура в м. Delhi', fontsize=15)
plt.show()
```

Додавання легенди

```python
plt.plot(month_climate.index, month_climate.meantemp, label='day temperature')
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Температура в м. Delhi', fontsize=15)
plt.legend()
plt.show()
```

Визначення розміру діаграми

```python
plt.figure(figsize=(10, 5))
plt.plot(month_climate.index, month_climate.meantemp, label='day temperature')
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Температура в м. Delhi', fontsize=15)
plt.legend()
plt.grid()
plt.show()
```

```python
from pandas import Timestamp

plt.figure(figsize=(10, 5))
plt.plot(month_climate.index, month_climate.meantemp, label='day temperature')
plt.xlabel('Дата', color='midnightblue', fontsize=20)
plt.ylabel('Температура', color='midnightblue', fontsize=20)
plt.title('Температура в м. Delhi', fontsize=25)
plt.text(Timestamp("2016-06-30"), month_climate.loc['2016-06-30']['meantemp'],
         'Максимум', color="black", fontsize=15)
plt.legend()
plt.grid()
plt.show()
```

Створення підобластей https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

```python
# climate_df_2015
# climate_df_2016

fig, axs = plt.subplots(2, 1, figsize=(10, 5))

axs[0].plot(climate_df_2015.index, climate_df_2015.meantemp, label='day temperature')
axs[1].plot(climate_df_2016.index, climate_df_2016.meantemp, label='day temperature')

axs[0].set_title('2015', fontsize=10)
axs[1].set_title('2016', fontsize=10)

plt.legend()
plt.grid()

fig.suptitle('Температура в м. Delhi', fontsize=15)

plt.legend()
plt.grid()
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 5))

axs[0].plot(climate_df_2015.index, climate_df_2015.meantemp,
            label='day temperature', linestyle=":", color="#061358")
axs[1].plot(climate_df_2016.index, climate_df_2016.meantemp,
            label='day temperature', linestyle="--", color="#FF5733")

axs[0].set_title('2015', fontsize=10)
axs[1].set_title('2016', fontsize=10)

plt.legend()
plt.grid()

fig.suptitle('Температура в м. Delhi', fontsize=15)

plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(26, 18))
plt.plot(climate_df_2015.index, climate_df_2015.meantemp,
            label='day temperature', linestyle=":",
         color="#061358", linewidth=2, marker="^")
plt.plot(climate_df_2015.index, climate_df_2015.humidity,
            label='day humidity', linestyle="--",
         color="#FF5733", linewidth=2, marker="D")

plt.title('Температура/вологість в м. Delhi', fontsize=15)

plt.legend()
plt.grid()
plt.show()
```

Візуалізація всіх даних у циклі

```python
plt.figure(figsize=(26, 18))
colors = ['#FF7F50', '#DDA0DD', '#66CDAA', '#BC8F8F']
for i, j in enumerate(climate_df.columns):
  plt.subplot(len(climate_df.columns) + 1, 1, i + 1)
  plt.plot(climate_df[j], color=colors[i]);
  plt.ylabel(j, fontsize=16)
  plt.grid()
plt.xlabel('Date', fontsize=16)
plt.show()
```

##Тривимірні графіки

###Лінійний графік

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

theta_max = 8 * np.pi
n = 1000
theta = np.linspace(0, theta_max, n)
x = theta
z = np.sin(theta)
y = np.cos(theta)
ax.plot(x, y, z, "g")

plt.show()
```

###Діаграма розсіювання

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

x = sales_df.Marital_Status
z = sales_df.Occupation
y = sales_df.Purchase

ax.scatter(x, y, z)

plt.show()
```

```python
iris_df = pd.read_csv("/drive/My Drive/!Kafedra/GOIT/mod2/Iris.csv")
iris_df
```

```python
iris_df['Species'].unique()
```

```python
enc_df = iris_df.replace({'Species': {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica':2}})
enc_df
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

x = iris_df.PetalLengthCm
z = iris_df.PetalWidthCm
y = iris_df.SepalLengthCm
marker = enc_df.Species
ax.scatter(x, y, z, c=marker)

ax.set_xlabel('PetalLengthCm Label')
ax.set_ylabel('PetalWidthCm Label')
ax.set_zlabel('SepalLengthCm Label')

plt.title("Розподіл за класами", fontsize=15)
plt.show()
```

###Каркасна поверхня

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

grid = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(grid, grid)
z = x ** 2 * y ** 2 + 2

ax.plot_wireframe(x, y, z)

plt.show()
```

###Поверхня

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

grid = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(grid, grid)
z = x ** 2 * y ** 2 + 2

ax.plot_surface(x, y, z)

plt.show()
```
