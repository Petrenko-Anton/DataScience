## Поняття про регресію

<pre>
<b>Регресійний аналіз</b> — розділ математичної статистики, присвячений методам
аналізу залежності однієї величини від іншої. На відміну від кореляційного
аналізу не з'ясовує чи істотний зв'язок, а займається пошуком моделі цього
зв'язку, вираженої у функції регресії.

Регресійний аналіз використовується в тому випадку, якщо відношення між
змінними можуть бути виражені кількісно у виді деякої комбінації цих змінних.
Отримана комбінація використовується для передбачення значення, що може
приймати цільова (залежна) змінна, яка обчислюється на заданому наборі значень
вхідних (незалежних) змінних.

</pre>

```python
from random import uniform, sample
def e(w):
    return w**4 - 27*w^2+14*w + 120
```

```python
def de(w):
    return 4*w**3 - 54*w + 14
```

```python
def gd(w_0, iterations, gamma):
    w_i = w_0
    for i in range(iterations):
        w_i = w_i - gamma*de(w_i)
        print('Iteration {} optimal w {}'.format(i, w_i))
    return w_i
```

```python
def gd_random(w_0, iterations):
    w_i = w_0
    gamma_list = [0.1, 0.01, 0.001, 0.0001, 0.00001]
    for i in range(iterations):
        # gamma = sample(gamma_list,1)[0]
        gamma = uniform(0.0001, 0.01)
        w_i = w_i - gamma*de(w_i)
        print('Iteration {} optimal w {}'.format(i, w_i))
    return w_i
```

##Приклад оптимізації двовимірної задачі методом градієнтного спуску


```python
def y_hat(w, x_val):
    return w[1]*x_val + w[0]
```

```python
def de(x, y, w, ind):
    m = len(x)
    error = [y_hat(w, x[i]) - y[i] for i in range(len(x))]
    if ind == 0:
        res = 1/m*(sum(error))
    if ind == 1:
        res = 1/m*(sum([x[i]*error[i] for i in range(len(x))]))
    return res
```

```python
def gd2(x, y, w_0, iterations, gamma):
    w_i = w_0
    for i in range(iterations):
        w_p = w_i
        w_i = [w_p[0] - gamma*de(x, y, w_p, 0), w_p[1] - gamma*de(x, y, w_p, 1)]
        print('Iteration {} optimal w {}'.format(i, w_i))
    return w_i
```

```python
x = [1.2, 2.5, 3.1, 4.3, 5.7]
y = [5.4, 8, 9.2, 11.6, 14.4]
```

```python
gd2(x, y, [3.3, 2.2], 1000, 0.001)
```

Або у формі векторів
\begin{align}
       \bar{y} = h_\omega(\mathbf{x})=\mathbf{\omega}^{T} \cdot \mathbf{x}
\end{align}
*   $\mathbf{\omega}$ - вектор параметрів моделі
*   $\mathbf{\omega}^T $ - транспонований вектор ознак
*   $\mathbf{x}$ - вектор ознак
*   $\mathbf{\omega}^{T} \cdot \mathbf{x}$ - скалярний добуток векторів


При навчанні моделі регресії розв'язується задача оптимізації функції помилок:
\begin{align}
       MSE(\mathbf{x},h_\omega) = \frac{1}{m} \sum_{i=1}^m (\mathbf{\omega}^{T} \cdot \mathbf{x}^{(i)}-y^{(i)})^2   
\end{align}
\begin{align}
       MSE->min  
\end{align}


Вибірка ознак згідно з результатами описового аналізу набору даних MPG. X - ознаки, Y - цільова змінна.

```python

def data_scale(data, scaler_type='minmax'):
    from sklearn.preprocessing import MinMaxScaler
    from sklearn.preprocessing import StandardScaler
    from sklearn.preprocessing import Normalizer
    if scaler_type == 'minmax':
        scaler = MinMaxScaler()
    if scaler_type == 'std':
        scaler = StandardScaler()
    if scaler_type == 'norm':
        scaler = preprocessing.Normalizer()

    scaler.fit(data)
    res = scaler.transform(data)
    return res
```
