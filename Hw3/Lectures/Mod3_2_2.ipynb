##Поняття про регуляризацію

Регуляризація моделі в машинному навчанні - це метод обмеження складності моделі. Зменшує помилку дисперсії

###Гребнева регресія


До функції помилок звичайної регресії додається штраф за великі значення коефіцієнтів моделі
\begin{align}
       ridge(\mathbf{x},h_\omega) = \frac{1}{m} \sum_{i=1}^m (\mathbf{\omega}^{T} \cdot \mathbf{x}^{(i)}-y^{(i)})^2 + \alpha \frac{1}{2} \sum_{i=1}^n {\omega_i}^2->min
\end{align}

Параметр $\alpha$ керує тим, як необхідно регуляризувати модель. Коли $\alpha=0$, гребнева регресія перетворюється в просту лінійну. При великих значеннях $\alpha$ всі параметри моделі $\omega_i$ стають близькими до 0.

```python
ridge = Ridge(alpha=0.9)
ridge.fit(x_train, y_train)
y_pred = ridge.predict(x_test)
```

### LASSO регресія
До функції помилок звичайної регресії додається штраф:
\begin{align}
       lasso(\mathbf{x},h_\omega) = \frac{1}{m} \sum_{i=1}^m (\mathbf{\omega}^{T} \cdot \mathbf{x}^{(i)}-y^{(i)})^2 + \alpha \sum_{i=1}^n \mid{\omega_i}\mid \to \min
\end{align}

```python
lasso = Lasso(alpha=0.001)
lasso.fit(x_train, y_train)
y_pred = lasso.predict(x_test)
```

###ElasticNet регресія

Еластична мережа є середнім між гребневою регресією та лассо-регресією. Керується ця суміш двох варіантів регуляризації параметром $r$
\begin{align}
       elastic(\mathbf{x},h_\omega) = \frac{1}{m} \sum_{i=1}^m (\mathbf{\omega}^{T} \cdot \mathbf{x}^{(i)}-y^{(i)})^2 + r \alpha \sum_{i=1}^n \mid{\omega_i}\mid + \frac{1-r}{2} \sum_{i=1}^n {\omega_i}^2 ->min
\end{align}


```python
elastic = ElasticNet(alpha=0.01, l1_ratio=0.3)
elastic.fit(x_train,y_train)
y_pred = elastic.predict(x_test)
```

##Налаштування гіперпараметрів

###Поняття про гіперпараметри

Гіперпараметри моделі - це параметри самої моделі, які керують процесом навчання.

Наприклад, у моделі ElasticNet, такими гіперпараметрами є $r$ та $\alpha$:

\begin{align}
       elastic(\mathbf{x},h_\omega) = \frac{1}{m} \sum_{i=1}^m (\mathbf{\omega}^{T} \cdot \mathbf{x}^{(i)}-y^{(i)})^2 + r \alpha \sum_{i=1}^n \mid{\omega_i}\mid + \frac{1-r}{2} \sum_{i=1}^n {\omega_i}^2 ->min
\end{align}

Гіперпараметри повинні бути визначені перед навчанням.



При цьому, $\omega_i$ є параметрами моделі, які визначаються під час навчання

Пошук оптимальних гіперпараметрів може призвести до суттєвого прокращення якості моделі.
(https://uk.wikipedia.org/wiki/%D0%93%D1%96%D0%BF%D0%B5%D1%80%D0%BF%D0%B0%D1%80%D0%B0%D0%BC%D0%B5%D1%82%D1%80_(%D0%BC%D0%B0%D1%88%D0%B8%D0%BD%D0%BD%D0%B5_%D0%BD%D0%B0%D0%B2%D1%87%D0%B0%D0%BD%D0%BD%D1%8F))

````python

def data_scale(data, scaler_type='minmax'):
    from sklearn.preprocessing import MinMaxScaler
    from sklearn.preprocessing import StandardScaler
    from sklearn.preprocessing import Normalizer
    if scaler_type == 'minmax':
        scaler = MinMaxScaler()
    if scaler_type == 'std':
        scaler = StandardScaler()
    if scaler_type == 'norm':
        scaler = preprocessing.Normalizer()

    scaler.fit(data)
    res = scaler.transform(data)
    return res
````
